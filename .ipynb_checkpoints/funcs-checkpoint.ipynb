```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
meta=yf.Ticker("META")
meta1=meta.history(period="6mo")
meta1.to_csv('Meta.csv')
data = pd.read_csv("Meta.csv")
dur = data['Date'].tolist()
close=data['Close'].tolist()
dur2=dur[0:108]
#dur20=dur[len(dur):len(dur)-20:1]

"""def ema(days):
    
    if days==1:
        Ema=1
    else:
        mult=2/(days-1)
        EMA = data['Close'] * mult + float(ema(days-1)) * (1-mult)
        return EMA
#year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
#unemployment_rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]
  
plt.plot(dur, ema(20), color='red', marker='o')
plt.title('unemployment rate vs year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('unemployment rate', fontsize=14)
plt.grid(True)
plt.show()"""
#arr = [1, 2, 3, 7, 9]
window_size = 20
  
i = 0
# Initialize an empty list to store moving averages
moving_averages = []
  
# Loop through the array to consider
# every window of size 3
while i < len(dur) - window_size + 1:
    
    # Store elements from i to i+window_size
    # in list to get the current window
    window = close[i : i + window_size]
   # print(window)
    # Calculate the average of current window
    window_average = sum(window) / window_size
      
    # Store the average of current
    # window in moving average list
    moving_averages.append(window_average)
      
    # Shift window to right by one position
    i += 1
  
#print(moving_averages)

plt.plot(dur2, moving_averages, color='red')
plt.title('date vs sma', fontsize=14)
plt.xlabel('date', fontsize=14)
plt.ylabel('sma', fontsize=14)
plt.grid(True)
plt.show()
```
